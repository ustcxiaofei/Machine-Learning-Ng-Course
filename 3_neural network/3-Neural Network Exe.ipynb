**定义导入数据函数**

```python
def load_data(path,transpose=True):
    """
    导入原始数据，按照需求是否对原始数据矩阵进行转置操作
    """
    data = sio.loadmat(path)
    y = data.get('y')  # (5000,1)
    X = data.get('X')  # (5000,400)  原始数据中一行代表一个数据 还未进行转置
    
    if transpose:
        X = np.array([im.reshape((20, 20)).T for im in X])
        X = np.array([im.reshape(400) for im in X])
    return X,y
```

# 按照一般的方法来构建训练通用逻辑回归二分类器

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def cost(theta,X,y):
    return np.mean(-y*np.log(sigmoid(X@theta))-(1-y)*np.log(1-sigmoid(X@theta)))
```

```python
def gridient(theta,X,y):
    return (1/len(X))*X.T@(sigmoid(X@theta)-y)
```

```python
def regularized_cost(theta,X,y,l=1):
    thetaOneToN = theta[1:]
    regTerm = (l/(2*len(X)))*np.power(thetaOneToN,2).sum()
    
    return cost(theta,X,y)+regTerm
```

```python
def regularized_gridient(theta,X,y,l=1):
    thetaOneToN = theta[1:]
    regTheta = (l/len(X))*thetaOneToN
    regTerm = np.concatenate([np.array([0]),regTheta])
    
    return gridient(theta,X,y)+regTerm
```

```python
def logistic_regression(X,y,l=1):
    theta = np.zeros(X.shape[1])
    res = opt.minimize(fun=regularized_cost,x0=theta,args=(X,y),jac=regularized_gridient)
    final_theta = res.x
    return final_theta
```

```python
def predict(x,theta):
    prob = sigmoid(x@theta)
    return (prob>=0.5).astype(int)
```

```python
t0 = logistic_regression(X,y[0])
print(t0.shape)
y_pred = predict(X,t0)
print('accuracy:',np.mean(y[0]==y_pred))
```

# 训练K个二分类器，实现多分类

```python
k_theta = np.array([logistic_regression(X,y[k],l=1) for k in range(10)])
print(k_theta.shape)
```

**进行多分类预测  选取二分类中预测最大值作为该类的真实类**
* think about the shape of k_theta, now you are making $X\times\theta^T$
> $(5000, 401) \times (10, 401).T = (5000, 10)$
* after that, you run sigmoid to get probabilities and for each row, you find the highest prob as the answer

```python
pred_matrix = sigmoid(X@k_theta.T)
pred_y = np.argmax(pred_matrix,axis=1)
anser_y = raw_y.copy()
anser_y[anser_y==10]=0
print(classification_report(anser_y,pred_y))
```

# 神经网络实现多分类

**网络模型如下**
![网络模型](./Image/NN model.PNG)

```python
def load_weight(path):
    data = sio.loadmat(path)
    return data['Theta1'],data['Theta2']
```

```python
theta1,theta2 = load_weight("ex3weights.mat")
print(theta1.shape,theta2.shape)
```

**注意在NN中数据的表示形式与以前不一致，需要将一个样本表示为X的一列，注意查看变量的形状**

```python
X,y = load_data("ex3data1.mat",transpose=False)
print(X.shape)
plot_an_image(X[2000,:])
```

```python
X = X.T
X = np.insert(X,0,values=np.zeros(X.shape[1]),axis=0)
print(X.shape,y.shape)
```

**NN前向传输过程**

```python
a1 = X
z2 = theta1@a1
print(z2.shape)
```

```python
a2 = sigmoid(z2)
a2.shape
```

```python
a2 = np.insert(a2,0,values=np.ones(z2.shape[1]),axis=0)
print(a2.shape)
```

```python
z3 = theta2@a2
z3.shape
```

```python
a3 = sigmoid(z3)
a3.shape
```

**注意其中预测值寻找最大位置后需要加1才与原始标记值意义一致**

```python
y_pred = np.argmax(a3,axis=0)
y_pred = y_pred+1
y_pred.shape
```

```python
print("accuracy:",classification_report(y,y_pred=y_pred))
```

**通过以上方式寻找最大值作为预测分类，则不需要进行one-hot编码？**

**one-hot 编码**

```python
def one_hot_code(y):
    '''
    y (5000,1)
    '''
    y_one_hot = np.zeros((y.shape[0],10))
    for ite in range(y.shape[0]):
        one_idx = y[ite]
        if one_idx == 10:
            one_idx=0
        y_one_hot[ite,one_idx] = 1
    return y_one_hot
```

```python
y = one_hot_code(y)
print(y.shape)
```
