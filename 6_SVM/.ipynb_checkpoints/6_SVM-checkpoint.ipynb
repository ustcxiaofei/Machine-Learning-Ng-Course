**C**

**注意上面的代价函数中C等价于逻辑回归中的1/lamda**

* C较大，相当于lamda很小，代价函数会去尽量使得每个样本都被正确分类，可能导致过拟合
* C较小，相当于lamda很大，可能会导致欠拟合现象

**高斯核函数参数sigma的影响**
* sigma较大，核函数较宽，即会有更多的地标样本纳入考虑，可能导致欠拟合，低方差
* sigma较小，核函数较窄，即仅仅考虑很相邻的样本，会提高训练准确的，低偏差，容易导致过拟合

## 逻辑回归、SVM、NN选择的一般准则：

**n为特征数，m为训练样本数**

1. 如果相较于样本而言，特征要大许多，即训练集数据量不够支持我们训练一个复杂的非线性模型，我们选用逻辑回归模型或者不带核函数的支持向量机。

2. 如果特征数较小，而且样本量大小中等，例如n在 1-1000 之间，而m在10-10000之间，使用高斯核函数的支持向量机。

3. 如果n较小，而m较大，例如m在1-1000之间，而m大于50000，则使用支持向量机会非常慢，解决方案是创造、增加更多的特征，然后使用逻辑回归或不带核函数的支持向量机。

值得一提的是，神经网络在以上三种情况下都可能会有较好的表现，但是训练神经网络可能非常慢，选择支持向量机的原因主要在于它的代价函数是凸函数，不存在局部最小值。

**尝试C= 1情况下的分类结果（直接调用sklearn中的svm函数来进行SVM实现）**

```python
svmLinear1 = sklearn.svm.LinearSVC(C=1,loss="hinge")
svmLinear1.fit(data[["X1","X2"]],data["y"])
svmLinear1.score(data[["X1","X2"]],data["y"])
```

**sklearn中根据决策函数绘制决策边界**

```python
def plot_svc_decision_function(model,ax=None,plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    
    #get the range of x
    xlim = ax.get_xlim()
    #get the range of y
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1],30)
    y = np.linspace(ylim[0], ylim[1],30)
    Y,X = np.meshgrid(y,x)
    
    #xy列
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k', levels=[-1, 0, 1], alpha=0.5,linestyles=['--', '-', '--'])
    #ax.contour(X, Y, P, colors='k', levels=[0], alpha=0.5,linestyles=[ '-'])
    
    # plot support vectors
    if plot_support:
        pass
        #ax.scatter(model.support_vectors_[:, 0],model.support_vectors_[:, 1],s=300, linewidth=1, facecolors='none')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(data["X1"], data["X2"], c=data["y"], s=50, cmap='autumn')
plot_svc_decision_function(svmLinear1)
```

**尝试C=100时候的情况**

## 高斯核SVM

```python
# kernek function 高斯核函数
def gaussian_kernel(x1, x2, sigma):
    return np.exp(- np.power(x1 - x2, 2).sum() / (2 * (sigma ** 2)))
```

```python
x1 = np.array([1, 2, 1])
x2 = np.array([0, 4, -1])
sigma = 2

gaussian_kernel(x1, x2, sigma)
```

## 网格参数搜索

**徒手搜索**

对于第三个数据集，我们给出了训练和验证集，并且基于验证集性能为SVM模型找到最优超参数。 虽然我们可以使用scikit-learn的内置网格搜索来做到这一点，但是本着遵循练习的目的，我们将从头开始实现一个简单的网格搜索。

```python
raw_data = sio.loadmat('./data/ex6data3.mat') # 'X', 'y', 'yval', 'Xval']

X = raw_data['X']
Xval = raw_data['Xval']
y = raw_data['y'].ravel()
yval = raw_data['yval'].ravel()

C_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100]
gamma_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100]

best_score = 0
best_params = {'C': None, 'gamma': None}

for C in C_values:
    for gamma in gamma_values:
        svc = sklearn.svm.SVC(C=C, gamma=gamma)
        svc.fit(X, y)
        score = svc.score(Xval, yval)
        
        if score > best_score:
            best_score = score
            best_params['C'] = C
            best_params['gamma'] = gamma

best_score, best_params
```

**使用sklearn中的GridSearchCV进行网格搜索**

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
parameters = {'C': C_values, 'gamma': gamma_values}
svc = sklearn.svm.SVC()
clf = GridSearchCV(svc,parameters,n_jobs=-1)
clf.fit(X,y)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
ypred = clf.predict(Xval)
print(classification_report(yval, ypred))
```

>curiouly... they are not the same result. What?  

It turns out that **GridSearch** will appropriate part of data as CV and use it to find the best candidate.  
So the reason for different result is just that GridSearch here is just using part of **training data** to train because it need part of data as cv set

## 垃圾邮件分类

```python
from sklearn import svm
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
import scipy.io as sio
```

> I think the hard part is how to vecotrize emails.

**导入数据，查看其基本信息**

```python
spam_train = sio.loadmat('./data/spamTrain.mat')
spam_test = sio.loadmat('./data/spamTest.mat')
print(spam_train.keys())
print(spam_test.keys())
print(spam_train["X"].shape)
print(spam_test["Xtest"].shape)
```

```python
X = spam_train['X']
Xtest = spam_test['Xtest']
y = spam_train['y'].ravel()
ytest = spam_test['ytest'].ravel()

X.shape, y.shape, Xtest.shape, ytest.shape
```

每个文档被表示为了1899维的向量，表示某个单词是否出现在文档之中，练习任务在于构建分类器基于这些特征向量**一共1899个特征，每个特征用0或1来表示**判断是否为垃圾邮件

```python
svc = svm.SVC()
svc.fit(X, y)
print('Training accuracy = {0}%'.format(np.round(svc.score(X, y) * 100, 2)))
```

```python
print('Test accuracy = {0}%'.format(np.round(svc.score(Xtest, ytest) * 100, 2)))
pred = svc.predict(Xtest)
print(metrics.classification_report(ytest, pred))
```

**对比下逻辑回归**

```python
logit = LogisticRegression()
logit.fit(X, y)
```

```python
pred = logit.predict(Xtest)
print(metrics.classification_report(ytest, pred))
```

**结果发现SVC还没有LG效果好，而且LG运行速度交SVC要快很多**
